```python
import pandas as pd

# Fill missing values
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('No cast listed')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')

# Convert date
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract numeric and type from duration
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
df['duration_type'] = df['duration'].str.extract(r'([A-Za-z]+)')

# Split genres
df['genres'] = df['listed_in'].str.split(', ')

# Strip strings
df['title'] = df['title'].str.strip()
df['type'] = df['type'].str.strip()
df['description'] = df['description'].str.strip()

# Drop duplicates
df = df.drop_duplicates(subset=['show_id'])
```

```python
# Lowercase columns that are categorical or for text analysis
df['type'] = df['type'].str.lower()
df['director'] = df['director'].str.title()
df['country'] = df['country'].str.title()
df['rating'] = df['rating'].str.upper()


# Remove extra spaces in cast and genres
df['cast'] = df['cast'].apply(lambda x: ', '.join([c.strip() for c in x.split(',')]))
df['genres'] = df['genres'].apply(lambda lst: [g.strip() for g in lst])


# Add a flag if director or country was originally missing
df['unknown_director'] = df['director'].apply(lambda x: 1 if x == 'Unknown' else 0)
df['unknown_country'] = df['country'].apply(lambda x: 1 if x == 'Unknown' else 0)


# Replace rare or inconsistent ratings
valid_ratings = ['G', 'PG', 'PG-13', 'R', 'NC-17', 'TV-MA', 'TV-PG', 'TV-14', 'Unknown']
df['rating'] = df['rating'].apply(lambda x: x if x in valid_ratings else 'Unknown')


# Number of cast members
df['num_cast'] = df['cast'].apply(lambda x: len(x.split(', ')) if x != 'No cast listed' else 0)

# Year added (from date_added)
df['year_added'] = df['date_added'].dt.year

# Duration split into minutes or seasons
df['duration_min'] = df.apply(lambda row: row['duration_num'] if row['duration_type'] == 'min' else None, axis=1)
df['num_seasons'] = df.apply(lambda row: row['duration_num'] if str(row['duration_type']).lower() == 'season' else None, axis=1)


```

```python
# Fill missing date_added and derive year_added
df['date_added'] = df['date_added'].fillna(pd.Timestamp('1900-01-01'))
df['year_added'] = df['date_added'].dt.year

# Fill missing duration_num with median
df['duration_num'] = df['duration_num'].fillna(df['duration_num'].median())

# Correctly create duration_min and num_seasons
df['duration_min'] = df.apply(
    lambda row: row['duration_num'] if str(row['duration_type']).lower() == 'min' else None,
    axis=1
)
df['num_seasons'] = df.apply(
    lambda row: row['duration_num'] if str(row['duration_type']).lower() == 'season' else None,
    axis=1
)

# Check all non-text columns for NaN
print(df.select_dtypes(include=['float64', 'int64']).isnull().sum())

# Check text columns for NaN
print(df.select_dtypes(include=['object']).isnull().sum())
```
